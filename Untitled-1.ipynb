```python
import numpy as np
import matplotlib.pyplot as plt
from d6py import D6Solver
import copy
import matplotlib.ticker as ticker
import matplotlib.patches as mpatches
import matplotlib as mpl
```

```python
initial_energy = 1e3  # GeV
final_energy = 1e4    # GeV
```

```python

running_variable_names = [
    'g',
    'gp',
    'gs',
    'Lambda',
    'm2',
    'Theta',
    'Theta_p',
    'Theta_s',
    'Gu',
    'Gd',
    'Ge',
    'Cduql[0, 0, 0, 0]',
    'Cduql[0, 2, 2, 0]',
    'Cqu1[0, 2, 0, 2]',
    'Cqu1[2, 0, 2, 0]',
    'Cqu8[0, 2, 0, 2]',
    'Cqu8[2, 0, 2, 0]',
    'Cduue[0, 0, 0, 0]',
    'Cduue[0, 2, 2, 0]',
    'Cuu[0, 2, 0, 2]',
    'Cuu[2, 0, 2, 0]',
    'Cqqql[0, 0, 0, 0]',
    'Cqqql[0, 2, 2, 0]',
    'Cqq1[0, 2, 0, 2]',
    'Cqq1[2, 0, 2, 0]',
    'Cqq3[0, 2, 0, 2]',
    'Cqq3[2, 0, 2, 0]',
    'Cqque[0, 0, 0, 0]',
    'Cqque[0, 2, 2, 0]',
]
```

```python
# SM interactions at 1 TeV
SM_initial_values = {
    'g': 0.61812326,
    'gp': 0.34971682,
    'gs': 1.06231166,
    'Lambda': 0.14970634,
    'm2': 8374.09249516,
    'Theta': 0.0,
    'Theta_p': 0.0,
    'Theta_s': 0.0,

    'Gu': np.array([
        [1.07975495e-05, -0.00143886, 0.00528027],
        [2.49303694e-06, 0.0062274, -0.03741809],
        [4.07803368e-08, 0.00027053, 0.88939254]
    ]),

    'Gd': np.array([
        [2.37179788e-05, 2.51667205e-09, -2.48813715e-06],
        [1.23211993e-10, 0.00048444, 1.76324968e-05],
        [-2.79766464e-09, 4.04956892e-07, 0.02067499]
    ]),

    'Ge': np.array([
        [2.99726848e-06, 0.0, 0.0],
        [0.0, 0.00061975, 0.0],
        [0.0, 0.0, 0.01042301]
    ]),

    'Cduql[0, 0, 0, 0]': 6.57462e-32,
    'Cqque[0, 0, 0, 0]': 3.30579e-32,
    'Cqqql[0, 0, 0, 0]':6.57462e-32,
    'Cduue[0, 0, 0, 0]': 6.57462e-32,

    'Cqu1[0, 2, 0, 2]': 2e-8,
    'Cqu8[0, 2, 0, 2]': 4.1e-8,
    'Cqq1[0, 2, 0, 2]': 1.2e-7,
    'Cqq3[0, 2, 0, 2]': 1.2e-7,
    'Cuu[0, 2, 0, 2]': 6.8e-9,
}
```

#### ONLY SM INTERACTIONS

```python
d6 = D6Solver()
d6.INITIAL_ENERGY_SCALE = initial_energy
d6.FINAL_ENERGY_SCALE = final_energy
```

```python
for key, value in SM_initial_values.items():
    d6.Set_value(key, value)

ic1331 = 1e-10
d6.Set_value('Cduue[0,2,2,0]', ic1331)
d6.Set_value('Cduql[0,2,2,0]', ic1331)
d6.Set_value('Cqque[0,2,2,0]', ic1331)
d6.Set_value('Cqqql[0,2,2,0]', ic1331)
```

```python
d6.Initialize_running_variables(running_variable_names)
d6.load_beta_functions()
d6.Run()
```

```python
print('Cduql[0,0,0,0]:', d6.Get_values(initial_energy, 'Cduql[0,0,0,0]'), '->', d6.Get_values(final_energy, 'Cduql[0,0,0,0]'))
print('Cqque[0,0,0,0]:', d6.Get_values(initial_energy, 'Cqque[0,0,0,0]'), '->', d6.Get_values(final_energy, 'Cqque[0,0,0,0]'))
print('Cqqql[0,0,0,0]:', d6.Get_values(initial_energy, 'Cqqql[0,0,0,0]'), '->', d6.Get_values(final_energy, 'Cqqql[0,0,0,0]'))
print('Cduue[0,0,0,0]:', d6.Get_values(initial_energy, 'Cduue[0,0,0,0]'), '->', d6.Get_values(final_energy, 'Cduue[0,0,0,0]'))

print('Cqu1[0,2,0,2]:', d6.Get_values(initial_energy, 'Cqu1[0,2,0,2]'), '->', d6.Get_values(final_energy, 'Cqu1[0,2,0,2]'))
print('Cqu8[0,2,0,2]:', d6.Get_values(initial_energy, 'Cqu8[0,2,0,2]'), '->', d6.Get_values(final_energy, 'Cqu8[0,2,0,2]'))
print('Cqq1[0,2,0,2]:', d6.Get_values(initial_energy, 'Cqq1[0,2,0,2]'), '->', d6.Get_values(final_energy, 'Cqq1[0,2,0,2]'))
print('Cqq3[0,2,0,2]:', d6.Get_values(initial_energy, 'Cqq3[0,2,0,2]'), '->', d6.Get_values(final_energy, 'Cqq3[0,2,0,2]'))
print('Cuu[0,2,0,2]:', d6.Get_values(initial_energy, 'Cuu[0,2,0,2]'), '->', d6.Get_values(final_energy, 'Cuu[0,2,0,2]'))
```

Change Initial Conditions and Run() again

```python
# copy old solution
ic1331_list = [ic1331]
solutions_SM = []
solutions_SM.append(copy.deepcopy(d6.Solution))
```

```python

for ic1331 in [1e-16, 1e-22, 1e-28]:
    d6.Set_value('Cduue[0,2,2,0]', ic1331)
    d6.Set_value('Cduql[0,2,2,0]', ic1331)
    d6.Set_value('Cqque[0,2,2,0]', ic1331)
    d6.Set_value('Cqqql[0,2,2,0]', ic1331)

    d6.Initialize_running_variables(running_variable_names)

    d6.Run()
    solutions_SM.append(copy.deepcopy(d6.Solution))
    ic1331_list.append(ic1331)

print(ic1331_list)

scale_1331_list = 1/np.sqrt(np.array(ic1331_list))
```

#### WITH BNC-BNV MIXING TERMS

```python
d6_tilde = D6Solver(initial_energy=initial_energy, final_energy=final_energy)
for key, value in SM_initial_values.items():
    d6_tilde.Set_value(key, value)

ic1331 = 1e-10

d6_tilde.Set_value('Cduue[0,2,2,0]', ic1331)
d6_tilde.Set_value('Cduql[0,2,2,0]', ic1331)
d6_tilde.Set_value('Cqque[0,2,2,0]', ic1331)
d6_tilde.Set_value('Cqqql[0,2,2,0]', ic1331)

d6_tilde.Initialize_running_variables(running_variable_names)

d6_tilde.ADDITIONAL_CONSTANTS["mt"] = 177.76
d6_tilde.ADDITIONAL_CONSTANTS["mb"] = 4.18

# WE HAVE CHANGED THE BETA FUNCTIONS
d6_tilde.load_beta_functions(Beta_fns_file="beta_functions/BetaSMEFT_tilde.txt")
d6_tilde.Run()
```

```python
print('Cduql[0,0,0,0]:', d6_tilde.Get_values(1e3, 'Cduql[0,0,0,0]'), '->', d6_tilde.Get_values(1e4, 'Cduql[0,0,0,0]'))
print('Cqque[0,0,0,0]:', d6_tilde.Get_values(1e3, 'Cqque[0,0,0,0]'), '->', d6_tilde.Get_values(1e4, 'Cqque[0,0,0,0]'))
print('Cqqql[0,0,0,0]:', d6_tilde.Get_values(1e3, 'Cqqql[0,0,0,0]'), '->', d6_tilde.Get_values(1e4, 'Cqqql[0,0,0,0]'))
print('Cduue[0,0,0,0]:', d6_tilde.Get_values(1e3, 'Cduue[0,0,0,0]'), '->', d6_tilde.Get_values(1e4, 'Cduue[0,0,0,0]'))

print()

print('Cqu1[0,2,0,2]:', d6_tilde.Get_values(1e3, 'Cqu1[0,2,0,2]'), '->', d6_tilde.Get_values(1e4, 'Cqu1[0,2,0,2]'))
print('Cqu8[0,2,0,2]:', d6_tilde.Get_values(1e3, 'Cqu8[0,2,0,2]'), '->', d6_tilde.Get_values(1e4, 'Cqu8[0,2,0,2]'))
print('Cqq1[0,2,0,2]:', d6_tilde.Get_values(1e3, 'Cqq1[0,2,0,2]'), '->', d6_tilde.Get_values(1e4, 'Cqq1[0,2,0,2]'))
print('Cqq3[0,2,0,2]:', d6_tilde.Get_values(1e3, 'Cqq3[0,2,0,2]'), '->', d6_tilde.Get_values(1e4, 'Cqq3[0,2,0,2]'))
print('Cuu[0,2,0,2]:', d6_tilde.Get_values(1e3, 'Cuu[0,2,0,2]'), '->', d6_tilde.Get_values(1e4, 'Cuu[0,2,0,2]'))

print()

print('Cduue[0,2,2,0]:', d6_tilde.Get_values(1e3, 'Cduue[0,2,2,0]'), '->', d6_tilde.Get_values(1e4, 'Cduue[0,2,2,0]'))
print('Cduql[0,2,2,0]:', d6_tilde.Get_values(1e3, 'Cduql[0,2,2,0]'), '->', d6_tilde.Get_values(1e4, 'Cduql[0,2,2,0]'))
print('Cqque[0,2,2,0]:', d6_tilde.Get_values(1e3, 'Cqque[0,2,2,0]'), '->', d6_tilde.Get_values(1e4, 'Cqque[0,2,2,0]'))
print('Cqqql[0,2,2,0]:', d6_tilde.Get_values(1e3, 'Cqqql[0,2,2,0]'), '->', d6_tilde.Get_values(1e4, 'Cqqql[0,2,2,0]'))
```

Change Initial Conditions and Run() again

```python
# copy old solution
ic1331_list = [ic1331]
solutions_tilde = []
solutions_tilde.append(copy.deepcopy(d6_tilde.Solution))
```

```python

for ic1331 in [1e-16, 1e-22, 1e-28]:
    d6_tilde.Set_value('Cduue[0,2,2,0]', ic1331)
    d6_tilde.Set_value('Cduql[0,2,2,0]', ic1331)
    d6_tilde.Set_value('Cqque[0,2,2,0]', ic1331)
    d6_tilde.Set_value('Cqqql[0,2,2,0]', ic1331)

    d6_tilde.Initialize_running_variables(running_variable_names)

    d6_tilde.Run()
    solutions_tilde.append(copy.deepcopy(d6_tilde.Solution))
    ic1331_list.append(ic1331)

print(ic1331_list)

scale_1331_list = 1/np.sqrt(np.array(ic1331_list))
```

```python
variable_names = [
    'Cduql[0,0,0,0]',
    'Cqque[0,0,0,0]',
    'Cqqql[0,0,0,0]',
    'Cduue[0,0,0,0]',
]

latex_variable_names = [
    r'$C^{duql}_{1111}$',
    r'$C^{qque}_{1111}$',
    r'$C^{qqql}_{1111}$',
    r'$C^{duue}_{1111}$',
]
```

```python
mu = np.linspace(initial_energy, final_energy, 100)
for i in range(len(ic1331_list)):
    fig, axs = plt.subplots(2, 4, figsize=(12, 5))
    for j in range(len(variable_names)):
        axs[0, j].plot(mu, d6.Get_values(mu, variable_names[j], solution=solutions_SM[i]))
        axs[0, j].set_xscale('log')
        axs[0, j].set_xlabel('$\mu$ (GeV)')
        axs[0, j].set_title(latex_variable_names[j])

        axs[1, j].plot(mu, d6_tilde.Get_values(mu, variable_names[j], solution=solutions_tilde[i]), color='tab:orange', linestyle='dashed')
        axs[1, j].set_xscale('log')
        axs[1, j].set_xlabel('$\mu$ (GeV)')
        #axs[1, j].set_ylabel(latex_variable_names[j])

    axs[0, 0].set_ylabel('SM only')
    axs[1, 0].set_ylabel('with BNV-BNC mixing')

    plt.tight_layout()
    plt.savefig('results/figures/case1/Case1_running_set' + str(i+1) + '.pdf')
    plt.show()
```

```python
Lambda_ini = [1/np.sqrt(np.abs(d6.Get_values(initial_energy, name)[0])) for name in variable_names]
```

```python
Lambda_sm_only = []
for solution in solutions_SM:
    Lambda_sm_only.append([1/np.sqrt(np.abs(d6.Get_values(final_energy, name, solution=solution)[0])) for name in variable_names])
```

```python
Lambda_with_loop_list = []
for solution in solutions_tilde:
    Lambda_with_loop = [1/np.sqrt(np.abs(d6_tilde.Get_values(final_energy, name, solution=solution)[0])) for name in variable_names]
    Lambda_with_loop_list.append(Lambda_with_loop)
```

```python
width = 1
fig, axs = plt.subplots(2, 2, figsize=(8,8))
axs = axs.flatten()

for i in range(len(variable_names)):
    axs[i].set_title(f"{latex_variable_names[i]}")
    axs[i].axhline(Lambda_ini[i], linestyle='dotted', color="black")
    axs[i].text(1.1, Lambda_ini[i], " Without RGE", ha='left', va='bottom')
    for j in range(len(ic1331_list)):
        line1 = axs[i].bar((j+1)*2, Lambda_sm_only[j][i], width, color="tab:blue")
        line2 = axs[i].bar((j+1)*2 + width*0.75, Lambda_with_loop_list[j][i], width, color="tab:orange")
    axs[i].set_yscale('log')

    #axs[i].set_xticks(np.arange(2.5, 9.5, 2), [f"{scale:.0e}" for scale in scale_1331_list])
    #axs[i].set_xlabel("Scale of $C_{1331}$ (GeV)")
    axs[i].set_xticks(np.arange(2.5, 9.5, 2), [f"Set {i+1}" for i in range(len(ic1331_list))])

""" axs[0].set_ylim(1.5e15, 5.5e15)
axs[2].set_ylim(3e15, 5.7e15)
axs[3].set_ylim(3.8e15, 4.7e15) """
fig.legend(handles=[line1, line2], labels=['SM only', 'with BNV-BNC mixing'], ncols=2, loc=(0.35, 0.96))
fig.supylabel(r'$\frac{\Lambda}{\sqrt{c}}$ (GeV)')
plt.tight_layout(w_pad=1.0)
plt.savefig("results/figures/case1/Lambda_comparison_case1.pdf")
plt.show()
```

```python
# print all Lambda values
print(f"Running from {initial_energy} GeV -> {final_energy} GeV")
for i in range(len(variable_names)):
    print(f"{variable_names[i]}: ")
    print(f"Without RGE -> Lambda = {Lambda_ini[i]:.2e}, WC = {d6.Get_values(initial_energy, variable_names[i])[0]:.2e}")
    for j in range(len(ic1331_list)):
        print(f"   C1331 -> Lambda = {scale_1331_list[j]:.0e}, WC = {ic1331_list[j]:.0e}")
        print(f"     SM only -> Lambda = {Lambda_sm_only[j][i]:.2e}, WC = {d6.Get_values(final_energy, variable_names[i], solution=solutions_SM[j])[0]:.2e}")
        print(f"     with BNV-BNC mixing -> Lambda = {Lambda_with_loop_list[j][i]:.2e}, WC = {d6_tilde.Get_values(final_energy, variable_names[i], solution=solutions_tilde[j])[0]:.2e}")
        print("")
    print("\n")
    
```

```python
import os

# Directory where all .tex files will go
output_dir = os.path.join("results", "latex")
os.makedirs(output_dir, exist_ok=True)

# Set your current case name (change this per run)
case_name = "case1"
case_title = case_name.replace("_", " ").title()

output_path = os.path.join(output_dir, f"{case_name}.tex")

with open(output_path, "w") as f:
    f.write(r"% Auto-generated LaTeX results for " + case_name + "\n\n")
    f.write(r"\subsection{" + case_title + "}" + "\n\n")
    f.write(f"Running from {initial_energy:.1e} GeV to {final_energy:.1e} GeV\n\n")

    for i in range(len(variable_names)):
        name = variable_names[i]
        latex_name = latex_variable_names[i]

        # --- Table for C1331 values ---
        f.write(r"\begin{center}" + "\n")
        # --- Without RGE ---
        f.write(f"{latex_name}" + " (Without RGE): "
                + r"$\Lambda_0 = $" + f" {Lambda_ini[i]:.2e}, "
                + r"$\text{WC}_0 = $" + f" {d6.Get_values(initial_energy, name)[0]:.2e}" + "\n\n")

                # Table: now 9 columns (added Set column at the start)
        f.write(r"\begin{tabular}{|c|cc|cc|cc|c|c|}" + "\n")
        f.write(r"\toprule" + "\n")

        # --- Top-level headers ---
        f.write(
            r"\multirow{3}{*}{Set} & "
            r"\multicolumn{2}{c|}{\multirow{2}{*}{$C_{1331}^{IC}$ Input}} & "
            r"\multicolumn{4}{c|}{" + latex_name + r"} & "
            r"\multirow{3}{*}{$\Lambda_{\text{SM}} / \Lambda_{0}$} & "
            r"\multirow{3}{*}{$\Lambda_{\text{Mix}} / \Lambda_{0}$} \\"
            "\n"
        )
        f.write(r"\cmidrule{4-7}" + "\n")
        f.write(
            r"& & & \multicolumn{2}{c|}{SM Only} & \multicolumn{2}{c|}{BNV--BNC Mixing} & & \\"
            "\n"
        )
        f.write(r"\cmidrule{2-7}" + "\n")

        # --- Column labels row ---
        f.write(
            r"& $\Lambda$ & WC & "
            r"$\Lambda_{\text{SM}}$ & $\text{WC}_{\text{SM}}$ & "
            r"$\Lambda_{\text{Mix}}$ & $\text{WC}_{\text{Mix}}$ & "
            r" &  \\"
            "\n"
        )
        f.write(r"\midrule" + "\n")

        # --- Data rows ---
        for j in range(len(ic1331_list)):
            scale = scale_1331_list[j]
            C1331 = ic1331_list[j]

            Lambda_noRGE = Lambda_ini[i]  # base scale (no RGE)
            Lambda_sm = Lambda_sm_only[j][i]
            C1111_sm = d6.Get_values(final_energy, name, solution=solutions_SM[j])[0]
            Lambda_mix = Lambda_with_loop_list[j][i]
            C1111_mix = d6_tilde.Get_values(final_energy, name, solution=solutions_tilde[j])[0]

            ratio_sm = Lambda_sm / Lambda_noRGE if Lambda_noRGE != 0 else 0
            ratio_mix = Lambda_mix / Lambda_noRGE if Lambda_noRGE != 0 else 0

            f.write(
                f"Set {j+1} & "
                f"{scale:.0e} & {C1331:.0e} & "
                f"{Lambda_sm:.2e} & {C1111_sm:.2e} & "
                f"{Lambda_mix:.2e} & {C1111_mix:.2e} & "
                f"{ratio_sm:.2} & {ratio_mix:.2} \\\\"
                "\n"
            )


        f.write(r"\bottomrule" + "\n")
        f.write(r"\end{tabular}" + "\n\n")
        f.write(r"\end{center}" + "\n\n")
        f.write(r"\vspace{0.5em}" + "\n\n")

print(f" LaTeX file saved at: {output_path}")
```
